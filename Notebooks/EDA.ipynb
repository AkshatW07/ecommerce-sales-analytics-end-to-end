```python
import pandas as pd
```

```python
df = pd.read_csv(r"C:\Users\AKSHAT\Ecommerce-Analytics-Project\Data\flipkart_com-ecommerce.csv")
df.head()
```

```python
df = df.drop(columns=["uniq_id","crawl_timestamp","product_url","image"])
```

```python
# handling missing values
df["brand"] = df["brand"].fillna("unknown")
```

```python
## dropping rows where price in missing
df = df.dropna(subset=["retail_price","discounted_price"])
```

```python
df.isnull().sum()
```

```python
df = df.drop(columns=["product_specifications"])
```

```python
df = df.drop(columns=['description'])
```

```python
df.isnull().sum()
```

```python
df['category'] = df['product_category_tree'].str.replace('[\[\]"]', '', regex=True)
```

```python
df['category'] = df['category'].str.split('>>').str[0].str.strip()
```

```python
df.to_csv(r"C:\Users\AKSHAT\Ecommerce-Analytics-Project\Data/clean_flipkart.csv")
```

```python
### feature 1 how much discount is given
df["discount_amount"] = df["retail_price"] - df["discounted_price"]
df.head()
```

```python
df.isnull().sum()
```

```python
## feature 2 discount percentage
df["discount_percent"] = (df["discount_amount"] / df["retail_price"]) * 100
```

```python
df.head()
```

```python
## feature 3 price category
def price_category(price):
    if price < 1000:
        return "low"
    elif price < 5000:
        return "medium"
    else:
        return "high"

df["price_category"] = df["discounted_price"].apply(price_category)
```

```python
df.head()
```

```python
df['product_rating'] = pd.to_numeric(df['product_rating'], errors='coerce')
```

```python
## feature 4 rating category

def rating_category(rating):
    if rating >= 4:
        return 'Good'
    elif rating >= 2:
        return 'Average'
    else:
        return 'Poor'

df['rating_category'] = df['product_rating'].apply(rating_category)
        
```

```python
df.head()
```

```python

#spliting data 
products_df=df[["product_name","brand","category","product_rating"]]
pricing_df=df[["retail_price","discounted_price","discount_percent"]]
```

```python
products_df.reset_index(inplace=True)
products_df.rename(columns={"index":"product_id"},inplace=True)

pricing_df["product_id"] = products_df["product_id"]
```

```python
products_df.to_csv(r"C:\Users\AKSHAT\Ecommerce-Analytics-Project\Data\products.csv",index=False)
pricing_df.to_csv(r"C:\Users\AKSHAT\Ecommerce-Analytics-Project\Data\pricing.csv",index=False)
```
